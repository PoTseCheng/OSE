```python
import scipy.io
import pandas as pd
import os
```

```python
xlsx = pd.ExcelFile("DataAgriculture.xlsx")
xlsx_list=xlsx.sheet_names
xlsx_list.remove("Tabelle1")

#construct empty column names
column_names = ["a"]
df = pd.DataFrame(columns = column_names)

for sheet_name in xlsx_list:
    df_new = pd.read_excel(xlsx, sheet_name=sheet_name)
    df= pd.concat([df, df_new], axis=1)

del df["a"]   
add=df.columns.tolist()
df.loc[-1] = add 
df.index = df.index + 1  
df = df.sort_index()
df.columns=xlsx_list
```
